## 20.2: Finite Difference Approximating Derivatives



```python
import matplotlib.pyplot as plt
import numpy as np

def graph_function(x_data, y_data, x_label, y_label, title, color):
  
  plt.style.use("seaborn-deep")
  plt.plot(x_data, y_data, color)
  plt.title(title)
  plt.xlabel(x_label)
  plt.ylabel(y_label)



def diff_plot(x, y):
  step = 0.1
  aprox_diff = np.diff(y)/step
  x_prime = x[:-1:]

  graph_function(x_prime,aprox_diff,"","","","--")


plt.figure(figsize=(10,8))
x = np.arange(0, 4*np.pi, 0.1) 
diff_plot(x, np.sin(x))

graph_function(x[:-1:], np.cos(x[:-1]),'','','','-g')


```

In this example I have taken the some of the code refrenced in the book and make it a fucntion to graph the Finite Difference Aproximation of any givin function. In this example, iI have taken the derivative of $sin(x)$ and shown the actual and aproximate solutions.

## 20.3 Approximating High Order Derivatives

```python
import matplotlib.pyplot as plt
import numpy as np

def graph_function(x_data, y_data, x_label, y_label, title, color):
  
  plt.style.use("seaborn-deep")
  #
  plt.plot(x_data, y_data, color)
  plt.title(title)
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  #plt.show()

def diff_plot(x, y,n_th):
  step = 0.1
  derivative = y
  x_prime = x
  for i in range(0,n_th):
    derivative = np.diff(derivative)/step
    x_prime = x_prime[:-1:]

  graph_function(x_prime,derivative,"","","","--")



plt.figure(figsize=(10,8))

x = np.arange(0, 4*np.pi, 0.1) 

diff_plot(x, np.cos(x),2)

graph_function(x[:-1:], -np.cos(x[:-1]),'','','','-g')


```

I have modified the function from the previous example to calculate the nth aproximate derivative given a function that is differentiable that many times.
In the example above I have taken the 2nd derivative of $cos(x)$ which is $-cos(x)$ I have also ploted the exact solution.

20.4: Numerical Differentiation with Noise

```python
import matplotlib.pyplot as plt
import numpy as np


def graph_function(x_data, y_data, x_label, y_label, title, color):
  
  plt.style.use("seaborn-deep")
  #
  plt.plot(x_data, y_data, color)
  plt.title(title)
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  #plt.show()

def diff_plot_noise(x, y,omega,epsilon):
  step = 0.1

  y_noisy = y + epsilon*omega*np.cos(omega*x)
  graph_function(x,y_noisy,"","","","-r")


x = np.arange(0, 2*np.pi, 0.01)
y = np.cos(x)

plt.figure(figsize=(10,8))
diff_plot_noise(x,y,100,0.01)
graph_function(x,y,"","","","-b")
```
